```python
class Kendaraan:
    def __init__(self, merek):
        self.merek = merek
    
    def info(self):
        print("ini adalah kendaraan", self.merek)
        
class Mobil(Kendaraan):
    def __init__(self, merek, jumlah_pintu):
        super().__init__(merek)
        self.jumlah_pintu = jumlah_pintu
        
    def info(self):
        super().info()
        print("ini adalah mobil dengan", self.jumlah_pintu, "pintu")
        
class Motor(Kendaraan):
    def __init__(self, merek, CC):
        super().__init__(merek)
        self.CC = CC
    
    def info(self):
        super().info()
        print("Ini adalah motor dengan cubic centimeter silinder: ", self.CC)

mobil = Mobil("Daihatsu", 5)
mobil.info()

motor = Motor("Yamaha", 150)
motor.info()
```

```python
class Induk:
    def __init__(self):
        self.nama = "induk"
        
    def metode(self):
        print("Metode dari kelas Induk")
        
class Anak(Induk):
    def __init__(self):
        super().__init__()
        self.nama = "Anak"
        
    def metode(self):
        super().metode()
        print("Metode dari kelas Anak")
        
object = Anak()
print(object.nama)
object.metode()
```

```python
class Induk1:
    def metode_induk1(self):
        print("Metode dari kelas Induk1")
        
class Induk2:
    def metode_induk2(self):
        print("Metode dari kelas Induk2")
        
class Anak(Induk1, Induk2):
    def metode_anak(self):
        print("Metode dari kelas anak")
        
object = Anak()
object.metode_induk1()
object.metode_induk2()
object.metode_anak()
```
