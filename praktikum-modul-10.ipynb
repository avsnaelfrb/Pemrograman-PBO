```python
class Induk:
    def metode_induk(self):
        print("Metode dari kelas Induk")

class Anak1(Induk):
    def metode_anak1(self):
        print("Metode anak 1")
        
class Anak2(Induk):
    def metode_anak2(self):
        print("Metode anak 2")

objek1 = Anak1()
objek1.metode_induk();
objek1.metode_anak1();

objek2 = Anak2()
objek2.metode_induk();
objek2.metode_anak2();
```

```python
class Induk2:
    def metode_induk2(self):
        print("metode dari kelas induk 2")
    

class Anak1(Induk):
    def metode_anak1(self):
        print("Metode anak 1")
        
class Anak2(Induk, Induk2):
    def metode_anak2(self):
        print("Metode anak 2")
        
class Cucu(Anak1, Anak2):
    def metode_cucu(self):
        print("Metode dari kelas cucu")
        
objek3 = Cucu()
objek3.metode_induk()
objek3.metode_induk2()
objek3.metode_anak1()
objek3.metode_anak2()
objek3.metode_cucu()
```

```python
class Induk:
    def __init__(self):
        self._atribut_pribadi = 10
    
    def _metode_pribadi(self):
        print("ini adalah metode pribadi dari kelas Induk")
        
    def metode_publik(self):
        print("ini adalah metode publik dari kelas")
        self._metode_pribadi()
        print("Nilai atribut pribadi", self._atribut_pribadi)
        
class Anak(Induk):
    def __init__(self):
        super().__init__()
        
    def akses_dari_anak(self):
        self._metode_pribadi()
        print("Nilai atribut pribadi:", self._atribut_pribadi)

objek4 = Anak()
objek4.metode_publik()
objek4.akses_dari_anak()
```
