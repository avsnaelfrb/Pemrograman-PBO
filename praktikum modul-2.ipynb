```python
#2.9
teks = "saya sedang memodifikasi teks pada python"
teks2 = "harga \tcabai naik"

print(teks.capitalize())
print(teks.casefold())
print(teks2.center(50))
print(teks.count("a"))
print(teks.encode())
print(teks.endswith("python"))
print(teks2.expandtabs(5))
print(teks.find("panda"))
```
