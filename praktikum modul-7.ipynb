```python
#7.6
class Manusia:
    def __init__(ytta, nama, umur):
        ytta.nama = nama
        ytta.umur = umur
        
    def cetak(test):
        print("Halo nama saya: " + test.nama)
        
    def cetak2(test):
        return test.umur
        
person2 = Manusia("ronal", 90)
person2.cetak() 
person2.cetak2() 
```

```python
#7.7
class Mahasiswa:
    def __init__(self, nama, jurusan, angkatan):
        self.nama = nama
        self.jurusan = jurusan
        self.angkatan = angkatan
    
    def informasi(self):
        return f"Nama : {self.nama} \nJurusan : {self.jurusan} \nAngkatan : {self.angkatan}"
    
    def __str__(self):
        return f"Mahasiswa: {self.nama}"

Mahasiswa1 = Mahasiswa("hapis", "teknik informatika", 2023)
Mahasiswa2 = Mahasiswa("hakim", "teknik informatika", 2024)

print(Mahasiswa1.nama)
print(Mahasiswa2.jurusan)

print(Mahasiswa1.informasi())

print(str(Mahasiswa2))
```
