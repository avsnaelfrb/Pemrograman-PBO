```python
class Karyawan:
    def __init__(self, nama, gaji):
        self.__nama = nama
        self.__gaji = gaji

    def dapatkan_nama(self):
        return self.__nama

    def atur_gaji(self, gaji_baru):
        if gaji_baru > 0:
            self.__gaji = gaji_baru
            print(f"Gaji {self.__nama} berhasil diubah menjadi {self.__gaji}")
        else:
            print("Gaji harus bernilai positif.")

    def tampilkan_detail(self):
        print(f"Nama: {self.__nama}, Gaji: Rp{self.__gaji}")

karyawan1 = Karyawan("Budi", 5000000)

karyawan1.tampilkan_detail() 

karyawan1.atur_gaji(-300000) 
karyawan1.atur_gaji(-100)    
karyawan1.tampilkan_detail() 
```

```python
class Kendaraan:
    def __init__(self, brand):
        self.brand = brand
        print(f"Kendaraan brand {self.brand} dibuat.")

    def mulai_mesin(self):
        print(f"{self.brand} mesin dinyalakan. Siap berjalan")

class Mobil(Kendaraan):
    def __init__(self, brand, model):
        super().__init__(brand) 
        self.model = model
        print(f"Mobil {self.brand} model {self.model} siap.")
    
class Motor(Kendaraan):
    def __init__(self, brand, tipe):
        super().__init__(brand)
        self.tipe = tipe
        print(f"Motor {self.brand} tipe {self.tipe} siap.")
    

print("\n--- DEMONSTRASI ---")

mobil_baru = Mobil("Toyota", "Avanza")
mobil_baru.mulai_mesin() 

print("-" * 25)

motor_sport = Motor("Yamaha", "R15")
motor_sport.mulai_mesin()

print("-" * 25)
```

```python
class KartuKredit:
    def proses_pembayaran(self):
        print("KARTU KREDIT: Otorisasi dan potong batas kredit.")

class TransferBank:
    def proses_pembayaran(self):
        print("TRANSFER BANK: Hasilkan kode virtual account dan tunggu konfirmasi.")

class DompetDigital:
    def proses_pembayaran(self):
        print("DOMPET DIGITAL: Kirim notifikasi konfirmasi ke aplikasi pengguna.")

daftar_metode = [
    KartuKredit(),
    TransferBank(),
    DompetDigital()
]

print("--- Memulai Proses Pembayaran pada Berbagai Metode ---")

for metode in daftar_metode:
    metode.proses_pembayaran()
    print("-" * 15)

print("Semua transaksi berhasil diproses.")
```

```python
class Tiket:
    def __init__(self, harga_dasar):
        self.__harga_dasar = harga_dasar

    def dapatkan_harga_dasar(self):
        return self.__harga_dasar

    def hitung_harga_akhir(self):
        return self.__harga_dasar

class TiketReguler(Tiket):
    def __init__(self, harga_dasar):
        Tiket.__init__(self, harga_dasar)
        
    def hitung_harga_akhir(self):
        return self.dapatkan_harga_dasar()

class TiketVIP(Tiket):
    def __init__(self, harga_dasar, biaya_tambahan):
        Tiket.__init__(self, harga_dasar)
        self.biaya_tambahan = biaya_tambahan
        
    def hitung_harga_akhir(self):
        return self.dapatkan_harga_dasar() + self.biaya_tambahan

harga_standar = 45000
tambahan_vip = 25000

tiket_reg = TiketReguler(harga_standar)
tiket_vip = TiketVIP(harga_standar, tambahan_vip)

semua_tiket = [tiket_reg, tiket_vip]

print("--- DEMONSTRASI HARGA TIKET ---")

for tiket in semua_tiket:
    jenis = tiket.__class__.__name__
    
    harga = tiket.hitung_harga_akhir() 
    
    print(f"Jenis Tiket: {jenis} | Harga Akhir: Rp{harga:,.0f}")
```
